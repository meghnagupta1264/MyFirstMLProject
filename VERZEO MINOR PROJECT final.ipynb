OBJECTIVE: To expose the best combination for strategy games available in the appstore in order to get a good user rating                  (4.0/5.0 and above) To analyse the dataset, we will look at the relationship as below:

```python
#Importing libraries
import pandas as pd
import numpy as np

#Reading .csv file
games=pd.read_csv("appstore.csv")
games.head()

#Removing unnecessary coloumns
games.drop(['URL','ID','Subtitle','Icon URL','Description','Developer','Languages','Primary Genre'],axis=1,inplace=True)
games.head()
```

```python
#Renaming few coloumns to easier names

games.rename(columns={'Average User Rating':'Rating','User Rating Count':'Count','In-app Purchases':'Purchases','Age Rating':'Age'},inplace=True)
games.columns
games.head()
```

```python
#Sorting the dataset into descending order based on user rating

games.sort_values('Rating',ascending=False).head()
```

```python
#Dropping all rows with null values as ratings

games=games.dropna(subset=['Rating'])
games
```

```python
#Dropping all rows from dataset having less than 4.0 (good) rating

games[games.Rating>=4.0]
```

```python
#Replacing all null values in Purchases coloumns with 0(zero)

games['Purchases'].fillna(0, inplace = True)
games.head()
```

```python
#Replacing certain Genre groups as Strategy as every row of dataset has Genre- Games, Entertainment, Strategy

games["Genres"]= games["Genres"].replace('Games, Strategy', 'Strategy')
games["Genres"]= games["Genres"].replace('Games, Strategy, Entertainment', 'Strategy')
games
```

```python
#Replacing all commas with spaces for better understanding of Genres coloumn
games['Genres'] = games['Genres'].str.replace('Games,', '')

#Renaming coloumn
games['Genres'] = games['Genres'].str.replace('Role Playing', 'Role')

games.head(20)
```

```python
#First converting wherever games exist in genres like (Entertainment, Adventure)= (Adventure); i.e one main Genre

#Primary Genres
games.loc[games['Genres'].str.contains('Card'),'Genres']='Card'
games.loc[games['Genres'].str.contains('Family'),'Genres']='Family'
games.loc[games['Genres'].str.contains('Board'),'Genres']='Board'
games.loc[games['Genres'].str.contains('Puzzle'),'Genres']='Puzzle'
games.loc[games['Genres'].str.contains('Action'),'Genres']='Action'
games.loc[games['Genres'].str.contains('Trivia'),'Genres']='Trivia'
games.loc[games['Genres'].str.contains('Simulation'),'Genres']='Simulation'
games.loc[games['Genres'].str.contains('Casual'),'Genres']='Casual'

#secondary genres
games.loc[games['Genres'].str.contains('Word'),'Genres']='Word'
games.loc[games['Genres'].str.contains('Adventure'),'Genres']='Adventure'
games.loc[games['Genres'].str.contains('Role'),'Genres']='Role'
games.loc[games['Genres'].str.contains('Sports'),'Genres']='Sports'
games.loc[games['Genres'].str.contains('Lifestyle'),'Genres']='Lifestyle'
games.loc[games['Genres'].str.contains('Utilities'),'Genres']='Utilities'
games.loc[games['Genres'].str.contains('Education'),'Genres']='Education'
games.loc[games['Genres'].str.contains('News Action'),'Genres']='News Action'
games.loc[games['Genres'].str.contains('Racing'),'Genres']='Racing'
games.loc[games['Genres'].str.contains('Health & Fitness'),'Genres']='Health & Fitness'




games.head(60)

#here games with multiple categories have been classified into a single main category
```

```python
#Converting the secondary genres from main genres into primary genres for simpler classifictaion

games.loc[games['Genres'].str.contains('Adventure'),'Genres']='Action'
games.loc[games['Genres'].str.contains('Role'),'Genres']='Action'
games.loc[games['Genres'].str.contains('Sports'),'Genres']='Action'
games.loc[games['Genres'].str.contains('Lifestyle'),'Genres']='Casual'
games.loc[games['Genres'].str.contains('Utilities'),'Genres']='Board'
games.loc[games['Genres'].str.contains('Education'),'Genres']='Family'
games.loc[games['Genres'].str.contains('Word'),'Genres']='Casual'
games.loc[games['Genres'].str.contains('News Action'),'Genres']='Education'
games.loc[games['Genres'].str.contains('Racing'),'Genres']='Action'
games.loc[games['Genres'].str.contains('Health & Fitness'),'Genres']='Casual'




games
```

```python
#Removing all rows with ratings less than 4.0
games = games[games['Rating'] >= 4.0]
games
```

```python
games = games.sort_values('Rating',ascending=False)
games.head(20)
```

```python
#Importing libraries

import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import matplotlib.mlab as mlab
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cbook as cbook
import seaborn as sns
%matplotlib inline
```

```python
# HISTOGRAM: No of games per main Genre

f, axes = plt.subplots ()

x=['Puzzle','Action','Trivia','Family','Card','Board','Simulation','Casual','Strategy']
y = [games.Genres[(games['Genres']=='Puzzle')].count(),games.Genres[(games['Genres']=='Action')].count(),
     games.Genres[(games['Genres']=='Trivia')].count(),games.Genres[(games['Genres']=='Family')].count(),
     games.Genres[(games['Genres']=='Card')].count(),games.Genres[(games['Genres']=='Board')].count(),
     games.Genres[(games['Genres']=='Simulation')].count(),games.Genres[(games['Genres']=='Casual')].count(),
     games.Genres[(games['Genres']=='Strategy')].count()]
hist_plot1= sns.barplot(x,y)  
hist_plot1.set(xlabel='Types Of Genres',ylabel='Number of Games in each Genre')
    
```

```python
# HISTOGRAM: Average User Rating Per Genre

x=['Puzzle','Action','Trivia','Family','Card','Board','Simulation','Casual','Strategy']
y = [games.Rating[(games['Genres']=='Puzzle')].mean(),games.Rating[(games['Genres']=='Action')].mean(),
     games.Rating[(games['Genres']=='Trivia')].mean(),games.Rating[(games['Genres']=='Family')].mean(),
     games.Rating[(games['Genres']=='Card')].mean(),games.Rating[(games['Genres']=='Board')].mean(),
     games.Rating[(games['Genres']=='Simulation')].mean(),games.Rating[(games['Genres']=='Casual')].mean(),
     games.Rating[(games['Genres']=='Strategy')].mean()]

hist_plot2= sns.barplot(x,y)
hist_plot2.set(xlabel='Types Of Genres',ylabel='Average User Rating per Genre')
hist_plot2.set(ylim=(4,5))

#for denoting the exact height (y coordinate) for histogram bars

for p in hist_plot2.patches:
    hist_plot2.annotate(("%.2f" % p.get_height()), (p.get_x()*1, p.get_height()) )
    
    
```

```python
# HISTOGRAM: Average Game Size Per Genre

#for converting game sizes into MB

games['SizeMB'] = round(games['Size']/1000000,1)

#plotting histogram

x=['Puzzle','Action','Trivia','Family','Card','Board','Simulation','Casual','Strategy']
y = [games.SizeMB[(games['Genres']=='Puzzle')].mean(),games.SizeMB[(games['Genres']=='Action')].mean(),
     games.SizeMB[(games['Genres']=='Trivia')].mean(),games.SizeMB[(games['Genres']=='Family')].mean(),
     games.SizeMB[(games['Genres']=='Card')].mean(),games.SizeMB[(games['Genres']=='Board')].mean(),
     games.SizeMB[(games['Genres']=='Simulation')].mean(),games.SizeMB[(games['Genres']=='Casual')].mean(),
     games.SizeMB[(games['Genres']=='Strategy')].mean()]

hist_plot3= sns.barplot(x,y)
hist_plot3.set(xlabel='Types Of Genres',ylabel='Average Game Size per Genre')
hist_plot3.set(ylim=(0,250))

#for denoting the exact height (y coordinate) for histogram bars

for p in hist_plot3.patches:
    hist_plot3.annotate(("%.2f" % p.get_height()), (p.get_x()*1, p.get_height()) )
    
```

```python
#STRIP PLOT: To show relation of user rating per genre vs game size

strip_plot= sns.stripplot(y='SizeMB' , x='Genres', data=games,\
                hue='Rating',dodge=True)

plt.legend(title='Rating',bbox_to_anchor=(2,1))

strip_plot.set(ylim=(0, 1000))
strip_plot.set(xlim=(0,8.5))
strip_plot.set(xlabel='Rating',ylabel='Game Sizes in MB')
```

```python
# PLOT: Average User Rating Per Age group

x=['4+','9+','12+','17+']
y = [games.Rating[(games['Age']=='4+')].mean(),games.Rating[(games['Age']=='9+')].mean(),
     games.Rating[(games['Age']=='12+')].mean(),games.Rating[(games['Age']=='17+')].mean()]

hist_plot3= sns.barplot(x,y)
hist_plot3.set(xlabel='age Groups',ylabel='Average User Rating per Age Group')
hist_plot3.set(ylim=(4,5))

#for denoting the exact height (y coordinate) for histogram bars

for p in hist_plot3.patches:
    hist_plot3.annotate(("%.2f" % p.get_height()), (p.get_x()*1, p.get_height()) )
    
```

```python
#PIE CHART: To show the percentage of games in dataset based on age rating

games["Age"].value_counts().plot.pie(autopct="%1.1f", explode=[0,0,0,0], figsize=(4,4))
plt.title("Age Rating wise percentage of games")
plt.show()
```

```python
#Converting string to date

games['Original Release Date'] = pd.to_datetime(games['Original Release Date'])
games['Current Version Release Date'] = pd.to_datetime(games['Current Version Release Date'])
```

```python
#Splitting the year and month from dates and making separate coloumns

games["Release Year"] = games["Original Release Date"].map(lambda x: x.strftime("%Y"))
games["Release Month"] = games["Original Release Date"].map(lambda x: x.strftime("%m"))
games
```

```python
#PLOT: To show trend of no of games released over the years

f, axes = plt.subplots()
games.groupby("Release Year")["Name"].count().plot(ax=axes)
axes.set_ylabel("no of games")
axes.set_title("Number of Games released per year ")

plt.show()
```

```python
#PLOT: To show the trend of the number of games released per month

f, axes = plt.subplots()
games.groupby("Release Month")["Name"].count().plot(ax=axes)
axes.set_ylabel("No. of games")
axes.set_title('Number Of Games released per month ')

plt.show()
```

```python
#PLOT: Average User rating based on release years

plt.figure(figsize=(10,5))
sns.barplot(data=games, x="Release Year", y="Rating")
plt.title("Average User rating based on release years")
plt.show()
```

```python
#PIE CHART: Release Year wise percentage of games

games["Release Year"].value_counts().plot.pie(autopct="%1.1f%%", figsize=(7,8))
plt.title("Release Year wise percentage of games")
plt.show()
```

```python
#Renaming a coloumn

games.rename(columns={'Release Year':'Year'},inplace=True)
games.columns
```
